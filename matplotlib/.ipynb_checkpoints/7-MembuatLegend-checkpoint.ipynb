```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# rumus
def sinusGenerator(amplitudo,frekuensi,tAkhir,theta):
    t = np.arange(0,tAkhir,0.1)
    y = amplitudo * np.sin(2 * frekuensi * t + np.deg2rad(theta))
    return t,y

# generate
amplitudo = 1
t1, y1 = sinusGenerator(amplitudo,1,4,0)
t2, y2 = sinusGenerator(amplitudo,1,4,30)
t3, y3 = sinusGenerator(amplitudo,1,4,60)
t4, y4 = sinusGenerator(amplitudo,1,4,90)

# decorate
judul  = 'Grafik Sinusoidal \n'
xlabel = 'Waktu(detik)'
ylabel = 'Magnituda(cm)'

# raw string (simbol mtk) || dengan cara latex
rumus      = r'$ \mathcal{Y} = A.sin(2 \omega t + \theta$' + '\n'
parameter1 = r'$ A = $' + str(amplitudo) + ' cm, '
parameter2 = r'$ \omega = $' + str(frekuensi) + r' $\mathit{Hz}$' + ', '
parameter3  = r'$ \theta = $' + str(theta) + r'$^{o} $'

# bikin plot
plt.plot(t1, y1, label="sin(0)")
plt.plot(t2, y2, label="sin(30)")
plt.plot(t3, y3, label="sin(60)")
plt.plot(t4, y4, label="sin(90)")
plt.title(judul + rumus + parameter1 + parameter2 + parameter3)
plt.xlabel(xlabel)
plt.ylabel(ylabel)


# fungsi legend = meberitahu grafik apa 
# tipe pertama : langsung tampilkan terserah dia posisi mau dimana
# plt.legend()

# tipe kedua : bisa menyesuaikan lokasi legend
# plt.legend(loc="lower left")

# tipe tiga = bisa ditaruh di luar
# plt.legend(loc="lower left", bbox_to_anchor=(0,-0.5))

# tipe empat 
# simple nya = subplot di ambil di ax trus kita perkecil ukurannya
plt.figure(1) # container
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.7, box.height])
plt.legend(loc="lower left", bbox_to_anchor=(1,0.685))


plt.show()
```
