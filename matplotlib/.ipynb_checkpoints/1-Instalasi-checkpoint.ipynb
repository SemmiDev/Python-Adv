```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,4,5,6,7,8]

# inisialasi plot
plt.plot(x,y)
# tampilkan
plt.show()
```

```python
# membuat lingkaran
PI = np.pi
sudut = np.linspace(0,2*PI, 100)
radius = 5
x = radius * np.cos(sudut)
y = radius * np.sin(sudut)
plt.plot(x,y)
plt.show()
```
