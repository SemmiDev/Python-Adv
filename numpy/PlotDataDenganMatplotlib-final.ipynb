```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# pers garis :: y = 2x + 3

x1 = np.arange(1,10,1)
y1 = 2 * x1 + 3
plt.figure(1)
plt.plot(x1,y1)

# lingkaran
r = 5
sudut = np.linspace(0,2*np.pi,100)
x2 =  r * np.cos(sudut)
y2 =  r * np.sin(sudut)
plt.figure(2)
plt.plot(x2,y2)

plt.show()

```
