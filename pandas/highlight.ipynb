```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
x = [8,4,9]
y = [15,5,7]
z1 = [6,12,4]
z2 = [6,9,9]
plt.plot(x,y,'g',label='Slope1',linewidth=3)
plt.plot(z1,z2,'c',label='Slope2',linewidth=3)
plt.title('Matplotlib graph')
plt.ylabel('Profit')
plt.xlabel('Sales')
plt.legend()
plt.grid(True,color='k')
```

```python
plt.show()
```
